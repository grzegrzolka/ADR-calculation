# Import libraries

```python
import pandas as pd
```

# Import excel file and load to pandas dataframe

```python
file_name = "D-Value_A1_A2_Exempt_corrected.xlsm"
df = pd.read_excel(file_name, sheet_name = "D-Value_A1_A2_Exempted")
df = df.set_index("Radionuclide")
df.head()
```

# Input 

```python
isotope_1 = "Cs-137"
activity_1 = 1 #MBq
activity_1 = activity_1 / 1000000 #TBq

isotope_2 = "Am-241"
activity_2 = 400 #MBq
activity_2 = activity_2 / 1000000 #TBq

isotope_3 = "Sr-90"
activity_3 = 1 #MBq
activity_3 = activity_3 / 1000000 #TBq

isotope_4 = "Se-75"
activity_4 = 1 #MBq
activity_4 = activity_4 / 1000000 #TBq

mass_kg = 10
is_solid = True

activity_list = [activity_1, activity_2, activity_3, activity_4 ] #TBq
selected_isotopes = df.loc[[isotope_1, isotope_2, isotope_3, isotope_4]]
selected_isotopes ["Activity [TBq]"] = activity_list #TBq
selected_isotopes
```

# Process data

```python
#### Suma aktywnosci wszystkich izotopow
activity_sum = selected_isotopes ["Activity [TBq]"].sum()


# f(i) - fraction of isotope activity in mixture - add value to the table
selected_isotopes ["f(i) fraction of isotope activity in mixture"] = selected_isotopes ["Activity [TBq]"] / activity_sum
# calculate "f(i) / A1" and "f(i) / A2"
selected_isotopes ["f(i)/A1(i)"] =  selected_isotopes ["f(i) fraction of isotope activity in mixture"] / selected_isotopes ["A1 [TBq]"]
selected_isotopes ["f(i)/A2(i)"] =  selected_isotopes ["f(i) fraction of isotope activity in mixture"] / selected_isotopes ["A2 [TBq]"]

### A1 mixture calclulation 
# fraction sum
fi_A1_sum = selected_isotopes ["f(i)/A1(i)"].sum()
# A1 mixture [TBq] 
A1_mixture = 1/fi_A1_sum
# Multiple of A1
multiple_A1 = activity_sum / A1_mixture

### A2 mixture calclulation 
# fraction sum
fi_A2_sum = selected_isotopes ["f(i)/A2(i)"].sum()
# A1 mixture [TBq] 
A2_mixture = 1/fi_A2_sum
# Multiple of A1
multiple_A2 = activity_sum / A2_mixture

### Transport security threshold
selected_isotopes ["f(i)/T(i)"] = selected_isotopes ["f(i) fraction of isotope activity in mixture"] / selected_isotopes ["Transport Security Threshold [TBq]"]
# T mixture [TBq]
fi_T_sum = selected_isotopes ["f(i)/T(i)"].sum()
T_mixture = 1 / fi_T_sum
# A / T
A_T = activity_sum / T_mixture

### Specific activity
selected_isotopes ["Specific activity [kBq / kg]"] = selected_isotopes ["Activity [TBq]"] * 1000000000 /mass_kg
specific_act_sum = selected_isotopes ["Specific activity [kBq / kg]"].sum()
### f(i) Activity concentration limit (i)
selected_isotopes ["f(i) / Activity concenctration limit (i)"] = selected_isotopes ["f(i) fraction of isotope activity in mixture"] / selected_isotopes ["Activity concentration for exempt material (kBq/kg)"]
### Activity concentration limit mixture [kBq/kg]
# activity concentration sum
act_concent_limit_sum = selected_isotopes ["f(i) / Activity concenctration limit (i)"].sum()
act_concent_limit_mixture = 1 / act_concent_limit_sum

selected_isotopes
```

# Kwalifikacja

```python
## HIGH CONSEQUENCES MIXTURE DANGEROUS GOODS
if A_T < 1:
    print("Not dangerous goods.")
else:
    print("Dangerous goods.")

## If needed expected package special form
if activity_sum <= A1_mixture / 1000:
    print("Excepted package special form.")
else:
    print("Not excepted package special form.")

#################################################
### Expected package solid
if is_solid:
    if activity_sum <= A2_mixture / 1000:
        print("Excepted package solid.")
    else:
        print("Not excepted package solid.")

### Expected package liquid
if not is_solid:
    if activity_sum <= A2_mixture / 10000:
        print("Excepted package liquid.")
    else:
        print("Not excepted package liquid.")
#################################################

## LSA I
if specific_act_sum <= 30 * act_concent_limit_mixture:
    print("LSA I")
else:
    print("Not LSA I")

## LSA II Solid
if is_solid:
    print("is solid")
    if specific_act_sum * 0.000000000001 <= A2_mixture * 0.0001:
        print("LSA II Solid")
        print("Not LSA III")
    else:
        print("Not LSA II Solid")
        if specific_act_sum * 0.000000000001 <= A2_mixture * 0.002 and specific_act_sum * 0.000000000001 >= A2_mixture * 0.0001:
            print("LSA III")
        else:
            print("Not LSA III")
else:
    print("is liquid")
    if specific_act_sum * 0.000000000001 <= A2_mixture * 0.00001:
        print("LSA II Liquid")
        print("Not LSA III")
    else:
        print("Not LSA II Liquid")
        print("Not LSA III")

## LSA III
if activity_sum <= A1_mixture:
    print("Type-A package")
else:
    print("Not Type-A package")
print("Activity/A: ", activity_sum/A1_mixture)
```
