```python
import pandas as pd
```

```python
file_name = "D-Value_A1_A2_Exempt.xlsm"
```

```python
df = pd.read_excel(file_name, sheet_name = "D-Value_A1_A2_Exempted")
```

```python
df.head()
```

```python
df.columns
```

```python
working_df = df[["Radionuclide", 
                "D-Value [TBq]", 
                "A1 [TBq]", 
                "A2 [TBq]", 
                "Transport Security Threshold [TBq]",
                "High Activity Source Directive 2013/59/EURATOM  [TBq]",
                "D-Value [Bq] P2",
                "Activity concentration for exempt material (Bq/g)",
                "Activity limit for an exempt consignment (Bq) P1",
                "Desintegration [Years]"]]
working_df
```

```python
isotope_1 = "Cs-137"
activity_1 = 700000
activity_1 = activity_1 / 1000000

isotope_2 = "Am-241"
activity_2 = 800000
activity_2 = activity_2 / 1000000

isotope_3 = "Sr-90"
activity_3 = 250000
activity_3 = activity_3 / 1000000

mass_kg = 10

activity_list = [activity_1, activity_2, activity_3 ]
```

```python
working_df = working_df.set_index("Radionuclide")
working_df
```

```python
selected_isotopes = working_df.loc[[isotope_1, isotope_2, isotope_3]]
```

```python
selected_isotopes ["Activity [TBq]"] = activity_list 
selected_isotopes 
```

```python
selected_isotopes ["Isotope Activity Ratio"] = selected_isotopes ["Activity [TBq]"] /selected_isotopes ["Activity [TBq]"].sum()
selected_isotopes
```

```python
selected_isotopes ["Multiple A1"] = selected_isotopes ["Activity [TBq]"] / selected_isotopes ["A1 [TBq]"]
selected_isotopes ["Multiple A2"] = selected_isotopes ["Activity [TBq]"] / selected_isotopes ["A2 [TBq]"]
selected_isotopes
```

```python
sum_A1 = selected_isotopes ["Multiple A1"].sum()
sum_A2 = selected_isotopes ["Multiple A2"].sum()
print("sum of multiple A1", round(sum_A1, 4))
print("sum of multiple A2", round(sum_A2, 4))
```

```python
selected_isotopes ["Activity concentration in consignment (Bq/g)"] = selected_isotopes ["Activity [TBq]"] * 1000000000 / mass_kg
selected_isotopes
```

```python
selected_isotopes["LSA-1"] = selected_isotopes ["Activity concentration in consignment (Bq/g)"] / 30 / selected_isotopes ["Activity concentration for exempt material (Bq/g)"]
selected_isotopes
```

```python
lsa_1_sum = selected_isotopes["LSA-1"].sum()
if lsa_1_sum  <= 1:
    print("IS LSA-1 (", selected_isotopes["LSA-1"].sum(), ")")
else:
    print("IS NOT LSA-1 (", selected_isotopes["LSA-1"].sum(), ")")
```

```python
selected_isotopes["LSA-2"] = selected_isotopes ["Activity concentration in consignment (Bq/g)"] / selected_isotopes["A2 [TBq]"] / 100000000
selected_isotopes
```

```python
lsa_2_sum = selected_isotopes["LSA-2"].sum()
if lsa_2_sum <= 1:
    print("IS LSA-2 (", lsa_2_sum, ")")
else:
    print("IS LSA-3 (", lsa_2_sum, ")")
```

```python
selected_isotopes["Type-A Package"] =  selected_isotopes ["Activity [TBq]"] / selected_isotopes ["A1 [TBq]"]
selected_isotopes
```

```python
type_A_sum = selected_isotopes["Type-A Package"].sum()
if type_A_sum <= 1:
    print("TYPE-A (", type_A_sum, ")")
else:
    print("NOT TYPE-A (", type_A_sum, ")")
```
